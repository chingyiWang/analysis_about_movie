```python
import pandas as pd
import numpy as np
from glob import glob
from datetime import datetime, timedelta
from pandas import DataFrame
```

```python
#將2020資料concat起來
files = glob("C:/Users/ching/Desktop/practice/movie/boxoffice/全國電影票房2020年*.xlsx")
df_2020 = pd.concat([pd.read_excel(f) for f in files])
df_2020.reset_index(drop=True)
```

```python
#資料型別轉換
df_2020['上映日期']=pd.to_datetime(df_2020['上映日期'])
df_2020['週票數變動率']=df_2020['週票數變動率'].str.replace('%','').str.replace(',','').astype(float)

tofloat = ['銷售金額','累計銷售票數','累計銷售金額']
for i in tofloat:
    df_2020[i]=df_2020[i].str.replace(',','').astype(float)
```

```python
#存成csv
Result ='C:/Users/ching/Desktop/practice/movie/data/df_2020.csv'
df_2020.to_csv(Result, index=True )
print( '成功產出'+ Result)
```
